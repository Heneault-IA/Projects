## Fonctions

```python
def keep_columns(df):

    cols = ["product_name", "description", "product_category_tree", "image"]

    return df[cols]


def concatenate_name_desc(data):
    """Fonction qui assemble les noms et descriptions"""
    
    df = data.copy()
    
    df["text"] = df["product_name"] + " " + df["description"]

    df.drop(columns=["product_name", "description"], inplace=True)

    return df


def main_category(data):
    """Fonction qui récupère la catégorie principale de chaque produits"""

    df = data.copy()
    
    df["main_category"] = df.apply(
        lambda x: x["product_category_tree"].split(">>")[0][2:-1], axis=1)

    df.drop(columns="product_category_tree", inplace=True)

    return df


def print_sample(sub_df, n_samp=10):
    """affiche n_samp documents depuis un corpus"""

    samp = sub_df.sample(n_samp)

    for idx, ser in samp.iterrows():

        txt_1 = f"target: {ser['main_category']}\n"
        print(txt_1 + ser["text"] + "..." + "\n\n")


def display_tokens_info(tokens):
    """Affiche les infos sur un corpus"""

    print(f"nb tokens : {len(tokens)}, nb tokens unique : {len(set(tokens))}")
    print(tokens[:30])


def process_text(doc,
                 rejoin=False,
                 lem_or_stem="lem",
                 stop_words=None,
                 list_common_words=None,
                 min_len_word=3,
                 force_is_alpha=True,
                 eng_words=None):
    """
    function witch lower, tokenize, clear stopwords, clear common_tokens,
    keep only certain lenght of words, choose if you want alpha words,
    lemmatize or Stemmatize, and can keep only eng_words.

    positional arguments:
    ---------------------
    doc : str : the document (aka a text in str format) to process

    opt args:
    --------------------
    rejoin : bool : if True, return a string else return the list of tokens
    lemm_or_stemm : str : if lem do lemmentize else stemmentize
    list_common_word : list : a list of common words to exclude
    min_len_word : int : the minimum lenght of words to not exclude
    force_is_alpha : int : if 1, exclude all tokens with a numeric character
    eng_words : list : list of english words

    return:
    -------------------
    a string( if rejoin is True) or a list of tokens
    """

    # stop_words
    if not stop_words:
        stop_words = []

    # list_common_words
    if not list_common_words:
        list_common_words = []

    # lower
    doc = doc.lower().strip()

    # tokenize
    tokenizer = RegexpTokenizer(r"\w+")
    raw_tokens_list = tokenizer.tokenize(doc)

    # classics stopwords
    cleaned_tokens_list = [w for w in raw_tokens_list if w not in stop_words]

    # no common tokens
    non_common_tokens = [
        w for w in cleaned_tokens_list if w not in list_common_words
    ]

    # no more len words
    more_than_N = [w for w in non_common_tokens if len(w) >= min_len_word]

    # only alpha chars
    if force_is_alpha:
        alpha_tokens = [w for w in more_than_N if w.isalpha()]
    else:
        alpha_tokens = more_than_N

    # stem or lem
    if lem_or_stem == "lem":
        trans = WordNetLemmatizer()
        trans_text = [trans.lemmatize(i) for i in alpha_tokens]
    elif lem_or_stem == "stem":
        trans = PorterStemmer()
        trans_text = [trans.stem(i) for i in alpha_tokens]

    # in english
    if eng_words:
        final = [i for i in trans_text if i in eng_words]
    else:
        final = trans_text

    # manage return type
    if rejoin:
        return " ".join(final)
    else:
        return final


def total_text_processing(data):
    """ perform ou final cleaning """
    
    df = data.copy()
    
    stop_words = set(stopwords.words("english"))

    eng_words = [i.lower() for i in words.words()]
    lemmatizer = WordNetLemmatizer()
    eng_words_lemm = [lemmatizer.lemmatize(i) for i in eng_words]

    list_df = []

    for categ in df["main_category"].unique():
        list_df.append(df[df["main_category"] == categ])
    
    list_corpus = []
    for tmp_df in list_df:
        tmp_corpus = " ".join(tmp_df["text"])

        list_corpus.append(tmp_corpus)
        
    list_processed_corpus = []

    for i, corpus in enumerate(list_corpus):
        
        list_processed_corpus.append(process_text(corpus))
        
    list_most_frequent = []

    for corpus in list_processed_corpus:
        list_most_frequent.append(pd.Series(corpus).value_counts().head(20).index)

    list_most_frequent
    
    occurrences = {}

    # Comptage des occurrences de chaque mot
    for liste in list_most_frequent:
        for mot in liste:
            if mot in occurrences:
                occurrences[mot] += 1
            else:
                occurrences[mot] = 1

    list_common_words = []

    for mot, freq in occurrences.items():
        if freq >=3:
            list_common_words.append(mot)
    
    df["clean_text"] = df["text"].apply(lambda r: process_text(r, rejoin=True,
                           stop_words=stop_words,
                           list_common_words=list_common_words,
                           eng_words=eng_words_lemm))
    return df
```
