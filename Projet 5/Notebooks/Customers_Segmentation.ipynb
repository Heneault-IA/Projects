## Fonctions

```python
def bootstrap_sample(data, n_samples):
    """
    Création de bootstrap avec sélection aléatoire des indices
    
    Args:
        data: DataFrame dans lequel faire la sélection
        n_samples : nombre d'indices à récupérer
    """
    sample_indices = np.random.choice(len(data), size=n_samples, replace=True)
    return data.loc[sample_indices]
```

# Baselines
## Colonnes RFM
### Récupération des données

```python
# Création du DataFrame avec les données du RFM
X_rfm = data[["Recency", "Frequency", "Monetary"]].copy()
X_rfm.head()
```

```python
X_rfm.dtypes
```

```python
# Affichage de la distribution de chaque variable
fig, axes = plt.subplots(1,3, figsize=(15,5))

# Boucle sur les colonnes
for i, col in enumerate(X_rfm.columns):
    sns.histplot(X_rfm[col], ax=axes[i])
    axes[i].set_title(f"Distribution pour {col}")
```

### Préparation des données et recherche du k optimisé

```python
# Standardisation des données
std_rfm_scaler = StandardScaler()

X_rfm_scaled = std_rfm_scaler.fit_transform(X_rfm)

X_rfm_scaled
```

```python
# Choix de la range de nombre de clusters à tester
range_n_cluster = range(3,10)

# Initialisation du Kmeans
kmeans_rfm = KMeans(random_state=42)

# Création d'une liste avec les différentes métriques utilisées par le KElbowVisualizer
metrics = ["distortion", "silhouette", "calinski_harabasz"]

# Boucle sur chaque métrique
for score in metrics:
    
    # Initialisation du KElbowVisualizer
    elbow_visualizer = KElbowVisualizer(kmeans_rfm, k=range_n_cluster, metric=score)
    
    # Entrainement sur les données scalées
    elbow_visualizer.fit(X_rfm_scaled)
    
    # Affichage du graphique
    elbow_visualizer.poof()
```

### Optimisation du KMeans

```python
# Choix du nombre de clusters optimaux
n_clusters = 4

# Initialisation du model
optimised_rfm_kmeans = KMeans(n_clusters=n_clusters, random_state=42)

# Création de la pipeline
optimised_rfm_pipeline = make_pipeline(std_rfm_scaler, optimised_rfm_kmeans)

# Entrainement sur les données
optimised_rfm_pipeline.fit(X_rfm)

# Récupération des labels
labels = optimised_rfm_pipeline["kmeans"].labels_

# Récupération des centroïds
centroids = optimised_rfm_pipeline["kmeans"].cluster_centers_
```

```python
# Ajout de la colonne Labels dans le DataFrame
X_rfm["Labels"] = labels

X_rfm.head(10)
```

```python
# Création du DataFrame des centroïds
centroids_df = pd.DataFrame(
    std_rfm_scaler.inverse_transform(centroids),
    columns=["Recency", "Frequency", "Monetary"],
    index=[f"Cluster {i}" for i in range(len(centroids))])
centroids_df
```

```python
# 0 = clients Récents : pas bcp dépensés et pas bcp d'achats mais achats récents
# 1 = Clients perdus : longtemps pas fait d'achat et 1 seul et pas dépensés bcp
# 2 = Clients à recontacter : bcp dépensés 1 fois mais rien depuis longtemps
# 3 = Clients à continuer de fidéliser : plusieurs achats, mais pas forcément récemment et moins
```

### Affichage en graphiques

```python
# Récupération des noms des labels
rfm_labels = X_rfm["Labels"].unique()
# Création d'une liste pour les légendes des graphiques
legend_rfm_labels = [f"Cluster {i}" for i in rfm_labels]

# Boucle sur chaque colonne excepté les labels
for col in X_rfm.drop(columns="Labels", axis=1).columns:

    # Affichage du graphqiue de distribution par label pour chaque colonne en boxplot
    sns.boxplot(data=X_rfm, x="Labels", y=col)
    plt.xticks(rfm_labels, labels=legend_rfm_labels)
    plt.show()
```

```python
# Regroupement par labels pour calculer le nombre d'individus dans chaque cluster
group_cluster = X_rfm.groupby("Labels").count()["Recency"]

# Affichage du graphique avec le nombre d'inidvidus dans chaque cluster
plt.bar(group_cluster.index, group_cluster)
plt.xticks(group_cluster.index, labels=legend_rfm_labels)
plt.ylabel("Nombre de Clients")
plt.title("Distribution des clients par cluster")
plt.show()
```

```python
# Affihcage du graphique avec les pourcentage d'inidvidus dans chaque cluter
plt.pie(group_cluster, labels = legend_rfm_labels, autopct="%1.1f%%")
plt.title("Pourcentage global par cluster")
plt.show()
```

### Vérification de la stabilité

```python
# Récupération des données pour chercher la stabilité
X_rfm_stability = X_rfm.drop(columns="Labels", axis=1).copy()

# Initialisation du StandardScaler et du modèle KMeans
rfm_stability_std_scaler = StandardScaler()
kmeans_rfm_original = KMeans(n_clusters=4, random_state=42)

# Création de la pipeline
rfm_stability_pipeline = make_pipeline(rfm_stability_std_scaler, kmeans_rfm_original)
# Entrainement sur les données
rfm_stability_pipeline.fit(X_rfm_stability)

# Obtention des centroids et labels initiaux
original_rfm_centroids = rfm_stability_pipeline["kmeans"].cluster_centers_
original_rfm_labels = rfm_stability_pipeline["kmeans"].labels_

# Initialisation pour stocker l'indice de confiance des assignations de points
rfm_labels_confidence = []

# nombre d'itération de tests du KMeans
n_iterations=100

# Boucle le nombre d'itération choisie
for _ in range(n_iterations):
    
    # Entrainement du K-Means
    rfm_stability_pipeline.fit(X_rfm_stability) 
    
    # Récupération des nouveaux labels
    new_labels = rfm_stability_pipeline["kmeans"].labels_
    
    # Calcul de l'indice de confiance entre les affectations originales et nouvelles
    ari = adjusted_rand_score(original_rfm_labels, new_labels)
    rfm_labels_confidence.append(ari)

# Calcul de la stabilité des assignations de points par la moyenne
labels_stability = np.mean(rfm_labels_confidence)
print("Indice de Confiance des Assignations de Points :", labels_stability)
```

```python
sns.lineplot(rfm_labels_confidence)
plt.xlabel("Nombre de tests")
plt.ylabel("Score Rand")
plt.title("Evolution du score Rand en fontion du nombre de tests")
plt.show()
```

### Sauvegarde des résultats

```python
# Ajout de la colonne de Last_Order pour les simulations
X_rfm["Last_Order"] = data["Last_Order"]

X_rfm.head()
```

```python
# Sauvegarde du fichier au format csv
X_rfm.to_csv("../Data/RFM_Data.csv", index=False)
```

## Calcul du meilleur EPS

```python
# Initialisation des listes pour récupérer les 
outlier_percent = []
number_outliers = []
number_clusters = []
score_silhouette = []

# boucle sur les valeurs eps cherchées
for eps in np.linspace(0.1, 1, 10):
    
    # Create Model
    dbscan = DBSCAN(eps=eps)
    dbscan_pipeline = make_pipeline(StandardScaler(), dbscan)
    # Pipeline
    dbscan_pipeline.fit(X_DBSCAN)
    
    labels_dbscan = dbscan_pipeline["dbscan"].labels_
    # Count Number of Outliers
    number_outliers.append(np.sum(labels_dbscan == -1))
    
    # Count Number of Clusters
    n_clusters = len(set(labels_dbscan)) - (1 if -1 in labels_dbscan else 0)
    number_clusters.append(n_clusters)
    
    # Log percentage of points that are outliers
    perc_outliers = 100 * np.sum(labels_dbscan == -1) / len(labels_dbscan)
    outlier_percent.append(perc_outliers)
    
    # Ajout du score de silhouette à la liste
    score_silhouette.append(silhouette_score(X_DBSCAN, labels_dbscan))
```

```python
# Affichage du graphique du nombre d'Outliers en fonction de la valeur Epsilon 
plt.figure(figsize = (8,8), dpi = 100)
sns.lineplot(x=np.linspace(0.1, 1, 10),y=number_outliers)
plt.ylabel("Number of Points Classified as Outliers")
plt.xlabel("Epsilon Value")
plt.title("Evolution du Nombre d'Outliers en fonction de la valeur Epsilon")
plt.show()
```

```python
# Affichage du graphique du pourcentage d'Outliers en fonction de la valeur Epsilon 
plt.figure(figsize = (8,8), dpi = 100)
sns.lineplot(x=np.linspace(0.1, 1, 10),y=outlier_percent)
plt.ylabel("Percentage of Points Classified as Outliers")
plt.xlabel("Epsilon Value")
plt.title("Evolution du pourcentage d'Outliers en fonction de la valeur Epsilon")
plt.show()
```

```python
# Affichage du graphique du nombre de clusters en fonction de la valeur Epsilon 
plt.figure(figsize = (8,8), dpi = 100)
sns.lineplot(x=np.linspace(0.1, 1, 10),y=number_clusters)
plt.ylabel("Number of Clusters Found")
plt.xlabel("Epsilon Value")
plt.title("Evolution du nombre de clusters en fonction de la valeur Epsilon")
plt.show()
```
