```python
#Data
import pandas as pd
import numpy as np

# Stats
from scipy.stats import randint, uniform
import statsmodels.api as sm

# Pre-processing
from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import TargetEncoder

# Sparse Matrix
from scipy import sparse

# Pipeline
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import make_union

# Transformer
from sklearn.compose import ColumnTransformer
from sklearn.compose import make_column_transformer
from sklearn.compose import make_column_selector
from sklearn.compose import TransformedTargetRegressor

# Model selection
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate

# Modelisation
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
import lightgbm as lgb

# Metrics
from sklearn.metrics import r2_score, mean_absolute_error

# Feature importance
import shap
import tensorflow

# Graphics
import matplotlib.pyplot as plt
import seaborn as sns

# Warnings
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)

# Time
import time
```

## Data

```python
# Importation des données
data = pd.read_csv("../Data/CleanData.csv", index_col=0)
data.head()
```

# Baseline modèle
## Récupération des données

```python
# Colonnes utiles pour la baseline
baseline_columns = [
    "YearBuilt", "NumberofBuildings", "NumberofFloors", "PropertyGFATotal",
    "PropertyGFABuilding(s)", "BuildingType", "PrimaryPropertyType",
    "LargestPropertyUseType"
]

# Préparation des données 
X = data[baseline_columns]
y = data["TotalGHGEmissions"]
```

## Séparation en train set et test set

```python
# Séparation en jeu d'entraienement et de test
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=42)
```

## Création de pipelines

```python
# Création du ColumnTransformer avec un StandardScaler et OneHotEncoder
columns_transformer = make_column_transformer(
    (StandardScaler(), make_column_selector(dtype_exclude=object)),
    (OneHotEncoder(handle_unknown="infrequent_if_exist"),
     make_column_selector(dtype_include=object)))
```

```python
# Création des modèles de régression linéaire
linear_estimator = LinearRegression()
elastic_net_estimator = ElasticNet()
lasso_estimator = Lasso()
ridge_estimator = Ridge()
```

```python
# Création des Pipelines avec le ColumnTransformer et les modèles
linear_pipeline = make_pipeline(columns_transformer, linear_estimator)
elastic_net_pipeline = make_pipeline(columns_transformer,
                                     elastic_net_estimator)
lasso_pipeline = make_pipeline(columns_transformer, lasso_estimator)
ridge_pipeline = make_pipeline(columns_transformer, ridge_estimator)

pipelines_baseline = [
    linear_pipeline, elastic_net_pipeline, lasso_pipeline, ridge_pipeline
]
```

## Entrainement et calcul des scores pour chaque modèle

```python
# Création de la liste des noms de modèles
list_pipeline_names = [
    "Linear Regression", "Elastic Net", "Lasso Regression", "Ridge Regression"
]

# Création de la liste des scores
list_scores_baseline = {}

# Création de la validation croisée
cv = KFold(n_splits=4, shuffle=True, random_state=42)

# Boucle sur chaque pipeline
for i, pipe in enumerate(pipelines_baseline):

    # Récupération du nom du modèle
    name_pipe = list_pipeline_names[i]

    # Ajout de la clé avec le nom et initialisation de la liste
    list_scores_baseline[name_pipe] = []

    # Création de la GridSearch pour la validation croisée
    grid = GridSearchCV(pipe,
                        param_grid={},
                        cv=cv,
                        scoring=["r2", "neg_mean_absolute_error"],
                        refit="r2")

    # Entrainement
    grid.fit(X_train, y_train)

    # Récupération des scores
    cv_results = grid.cv_results_

    # Séparation en scores de temps d'entrainement, R2 et NMAE
    training_time = round(cv_results["mean_fit_time"][0], 4)
    score_r2 = round(cv_results["mean_test_r2"][0], 2)
    score_nmae = round(cv_results["mean_test_neg_mean_absolute_error"][0], 2)

    # Ajout des scores à la clé associée
    list_scores_baseline[name_pipe].append(training_time)
    list_scores_baseline[name_pipe].append(score_r2)
    list_scores_baseline[name_pipe].append(score_nmae)

    # Validation croisée pour le jeu de test sur 4 segments
    cross_validation = cross_validate(
        grid, X_test, y_test, scoring=["r2", "neg_mean_absolute_error"], cv=cv)

    # # Séparation en scores de temps de scoring, R2 et NMAE
    training_time = round(cross_validation["score_time"].mean(), 4)
    score_r2 = round(cross_validation["test_r2"].mean(), 2)
    score_nmae = round(cross_validation["test_neg_mean_absolute_error"].mean(),
                       2)

    # Ajout des scores à la clé associée
    list_scores_baseline[name_pipe].append(training_time)
    list_scores_baseline[name_pipe].append(score_r2)
    list_scores_baseline[name_pipe].append(score_nmae)
```

```python
# Création du DataFrame des scores avec transposition et tri sur le score R2 du jeu de test
df_scores_baseline = pd.DataFrame(list_scores_baseline,
                         index=[
                             "Training Time Train", "Score R2 Train",
                             "Score NMAE Train", "Score Time Test",
                             "Score R2 Test", "Score NMAE Test"
                         ]).T.sort_values("Score R2 Test", ascending=False)
# Affichage des scores
df_scores_baseline
```

```python
# Création de la liste des noms de modèles
list_pipeline_names = [
    "SVM Regression", "Extra Trees Regression", "Gradient Boost Regression",
    "Random Forest Regression", "XG Boost Regression", "Light GBM Regression"
]

# Création de la liste des scores
list_scores_non_linear = {}

# Boucle sur chaque pipeline
for i, pipe in enumerate(pipelines_non_linear):

    # Récupération du nom du modèle
    name_pipe = list_pipeline_names[i]

    # Ajout de la clé avec le nom et initialisation de la liste
    list_scores_non_linear[name_pipe] = []

    # Création de la GridSearch pour la validation croisée sur 4 segments
    grid = GridSearchCV(pipe,
                        param_grid={},
                        cv=cv,
                        scoring=["r2", "neg_mean_absolute_error"],
                        refit="r2")

    # Entrainement
    grid.fit(X_train, y_train)

    # Récupération des scores
    cv_results = grid.cv_results_

    # Séparation en scores de temps d'entrainement, R2 et NMAE
    training_time = round(cv_results["mean_fit_time"][0], 4)
    score_r2 = round(cv_results["mean_test_r2"][0], 2)
    score_nmae = round(cv_results["mean_test_neg_mean_absolute_error"][0], 2)

    # Ajout des scores à la clé associée
    list_scores_non_linear[name_pipe].append(training_time)
    list_scores_non_linear[name_pipe].append(score_r2)
    list_scores_non_linear[name_pipe].append(score_nmae)

    # Validation croisée pour le jeu de test sur 4 segments
    cross_validation = cross_validate(
        grid, X_test, y_test, scoring=["r2", "neg_mean_absolute_error"], cv=cv)

    # # Séparation en scores de temps de scoring, R2 et NMAE
    training_time = round(cross_validation["score_time"].mean(), 4)
    score_r2 = round(cross_validation["test_r2"].mean(), 2)
    score_nmae = round(cross_validation["test_neg_mean_absolute_error"].mean(),
                       2)

    # Ajout des scores à la clé associée
    list_scores_non_linear[name_pipe].append(training_time)
    list_scores_non_linear[name_pipe].append(score_r2)
    list_scores_non_linear[name_pipe].append(score_nmae)
```

## Entraînement et calcul des scores

```python
# Création de la liste des scores
list_scores_transformed_target = {}

# Boucle sur chaque pipeline
for i, pipe in enumerate(pipelines_transformed_target):

    # Récupération du nom du modèle
    name_pipe = list_pipeline_names[i]

    # Ajout de la clé avec le nom et initialisation de la liste
    list_scores_transformed_target[name_pipe] = []

    # Création de la GridSearch pour la validation croisée sur 4 segments
    grid = GridSearchCV(pipe,
                        param_grid={},
                        cv=cv,
                        scoring=["r2", "neg_mean_absolute_error"],
                        refit="r2")

    # Entrainement
    grid.fit(X_train, y_train)

    # Récupération des scores
    cv_results = grid.cv_results_

    # Séparation en scores de temps d'entrainement, R2 et NMAE
    training_time = round(cv_results["mean_fit_time"][0], 4)
    score_r2 = round(cv_results["mean_test_r2"][0], 2)
    score_nmae = round(cv_results["mean_test_neg_mean_absolute_error"][0], 2)

    # Ajout des scores à la clé associée
    list_scores_transformed_target[name_pipe].append(training_time)
    list_scores_transformed_target[name_pipe].append(score_r2)
    list_scores_transformed_target[name_pipe].append(score_nmae)

    # Validation croisée pour le jeu de test sur 4 segments
    cross_validation = cross_validate(
        grid, X_test, y_test, cv=cv, scoring=["r2", "neg_mean_absolute_error"])

    # # Séparation en scores de temps de scoring, R2 et NMAE
    training_time = round(cross_validation["score_time"].mean(), 4)
    score_r2 = round(cross_validation["test_r2"].mean(), 2)
    score_nmae = round(cross_validation["test_neg_mean_absolute_error"].mean(),
                       2)

    # Ajout des scores à la clé associée
    list_scores_transformed_target[name_pipe].append(training_time)
    list_scores_transformed_target[name_pipe].append(score_r2)
    list_scores_transformed_target[name_pipe].append(score_nmae)
```
